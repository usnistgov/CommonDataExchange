```python
import os
import glob

xmls = glob.glob(os.path.join("sampleData","*.xml"))
print(xmls)
```

# use xmlschema to validate the data with schema

```python
import xmlschema
sampleSchema = xmlschema.XMLSchema(os.path.join("sampleData","dilatometry.xsd"))


sampleSchema.is_valid(xmls[0])
```

# transforming data format between XML and JSON using DataModelDict library

```python
import libs.DataModelDict as DMD

text_file = open(xmls[0], "r")
xmlCotent_ = text_file.read()
 
#close file
text_file.close()



xmlCotent = DMD.DataModelDict(xmlCotent_)
#print(xmlCotent)
```

# DMD provide query function to find keyword

```python
print(xmlCotent.finds('sensorType'))
```

# transform XML to JSON

```python
import json

jsonCotent_ = json.loads(xmlCotent.json())  #indent=2
print(jsonCotent_)
```

# Transform new JSON back to a new XML

```python
jsonCotent = DMD.DataModelDict(jsonCotent_)

newXMLCotent_ = jsonCotent.xml(indent=4)
print(type(newXMLCotent_))


newXMLCotent = DMD.DataModelDict(newXMLCotent_)
print(type(newXMLCotent))

newXMLFile = os.path.join("sampleData","newXML.xml")
with open(newXMLFile, 'w') as f:
    newXMLCotent.xml(fp=f)
```

# validate new XML usign the same schema

```python
sampleSchema.is_valid(newXMLFile)
```
